```python
import os
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Make sure to download the NLTK resources
nltk.download('punkt')

def process_html_file(file_path):
    # Read and parse the HTML file
    with open(file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')

    # Extract text from the parsed HTML
    text = soup.get_text()

    # Tokenize the text into words (tokens)
    tokens = word_tokenize(text)

    # Tokenize the text into sentences
    sentences = sent_tokenize(text)

    # Return the token and sentence counts
    return len(tokens), len(sentences)

def process_folder(folder_path):
    total_tokens = 0
    total_sentences = 0

    # Loop over all HTML files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.html'):
            file_path = os.path.join(folder_path, file_name)
            tokens, sentences = process_html_file(file_path)

            print(f"File: {file_name}")
            print(f"Tokens: {tokens}")
            print(f"Sentences: {sentences}")
            print()

            # Update total counts
            total_tokens += tokens
            total_sentences += sentences

    # Print the total tokens and sentences
    print(f"Total Tokens: {total_tokens}")
    print(f"Total Sentences: {total_sentences}")

# Example usage: Update the folder path where your HTML files are located
folder_path = r'C:\Users\archishman vb\OneDrive\Desktop\raw'
process_folder(folder_path)
```

total no. of annotated tokens

```python
import json
import glob
import os

def extract_annotated_tokens(json_path):
    """
    Extract the annotated tokens from a JSON file.
    Returns the count of annotated tokens.
    :param json_path: Path to the JSON file
    """
    with open(json_path, 'r', encoding='utf-8') as json_file:
        content = json.load(json_file)

    annotated_tokens = 0

    # Loop through the annotations and extract the tokens
    for item in content:
        for annotation in item.get('annotations', []):
            for result in annotation.get('result', []):
                value = result.get('value', {})
                text = value.get('text', '')

                if text:
                    # Split the text into words (tokens)
                    words = text.split()
                    annotated_tokens += len(words)  # Count tokens

    return annotated_tokens

def calculate_total_annotated_tokens(input_folder):
    """
    Loop over all JSON files in a folder and calculate the total number of annotated tokens.
    :param input_folder: Folder containing JSON files
    """
    total_annotated_tokens = 0
    json_paths = glob.glob(os.path.join(input_folder, '*.json'))  # Get all JSON files

    for json_path in json_paths:
        tokens_in_file = extract_annotated_tokens(json_path)
        total_annotated_tokens += tokens_in_file

    print(f"Total number of annotated tokens: {total_annotated_tokens}")

# Example usage: Specify your input folder
input_folder = r'C:\Users\archishman vb\OneDrive\Desktop\json_files'
calculate_total_annotated_tokens(input_folder)
```

total number of annotated sentences

```python
import json
import glob
import os

def extract_annotated_text(json_path):
    """
    Extract all annotated text from the JSON and return a list of annotated texts.
    :param json_path: Path to the JSON file
    :return: List of annotated texts
    """
    with open(json_path, 'r', encoding='utf-8') as json_file:
        content = json.load(json_file)

    annotated_texts = []

    # Loop through the annotations and extract the text
    for item in content:
        for annotation in item.get('annotations', []):
            for result in annotation.get('result', []):
                value = result.get('value', {})
                text = value.get('text', '')

                if text:
                    annotated_texts.append(text)  # Collect the annotated text

    return annotated_texts

def is_sentence(text, min_length=5):
    """
    Check if the text is composed of a sentence based on a minimum length.
    :param text: Annotated text to check
    :param min_length: Minimum number of words to consider it a sentence
    :return: Boolean indicating if the text qualifies as a sentence
    """
    words = text.split()
    return len(words) >= min_length

def count_annotated_sentences(input_folder):
    """
    Loop over all JSON files in a folder and count the total number of annotated texts
    that qualify as sentences.
    :param input_folder: Folder containing JSON files
    """
    total_sentences = 0
    json_paths = glob.glob(os.path.join(input_folder, '*.json'))  # Get all JSON files

    for json_path in json_paths:
        annotated_texts = extract_annotated_text(json_path)  # Get all annotated texts

        # Check each annotated text to see if it qualifies as a sentence
        for text in annotated_texts:
            if is_sentence(text):
                total_sentences += 1

    print(f"Total number of annotated sentences: {total_sentences}")

# Example usage: Specify your input folder
input_folder = r'C:\Users\archishman vb\OneDrive\Desktop\json_files'
count_annotated_sentences(input_folder)
```
